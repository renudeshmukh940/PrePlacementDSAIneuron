**Question 1**
Given an integer array nums of 2n integers, group these integers into n pairs (a1, b1), (a2, b2),..., (an, bn) such that the sum of min(ai, bi) for all i is maximized. Return the maximized sum.

**Example 1:**
Input: nums = [1,4,3,2]
Output: 4

**Explanation:** All possible pairings (ignoring the ordering of elements) are:

1. (1, 4), (2, 3) -> min(1, 4) + min(2, 3) = 1 + 2 = 3
2. (1, 3), (2, 4) -> min(1, 3) + min(2, 4) = 1 + 2 = 3
3. (1, 2), (3, 4) -> min(1, 2) + min(3, 4) = 1 + 3 = 4
So the maximum possible sum is 4

`Approach` :
1. Sort the array `nums` in ascending order.
2. Initialize a variable `max_sum` to 0.
3. Iterate over the sorted array `nums` with a step size of 2 (since we need to group the numbers in pairs).
4. For each pair `(a, b)`, add the minimum of `a` and `b` to `max_sum`.
5. Return the value of `max_sum`.

**Time Complexity** : ` O(n log n)` 

**Space Complexity** : `O(1)` 

```python
def array_pair_sum(nums):
    nums.sort()  
    max_sum = 0
    for i in range(0, len(nums), 2):
        max_sum += min(nums[i], nums[i+1])
    return max_sum
nums = [1, 4, 3, 2]
result = array_pair_sum(nums)
print(result)
```

**Question 2**
Alice has n candies, where the ith candy is of type candyType[i]. Alice noticed that she started to gain weight, so she visited a doctor. 

The doctor advised Alice to only eat n / 2 of the candies she has (n is always even). Alice likes her candies very much, and she wants to eat the maximum number of different types of candies while still following the doctor's advice. 

Given the integer array candyType of length n, return the maximum number of different types of candies she can eat if she only eats n / 2 of them.

**Example 1:**
Input: candyType = [1,1,2,2,3,3]
Output: 3

**Explanation:** Alice can only eat 6 / 2 = 3 candies. Since there are only 3 types, she can eat one of each type.

`Approach`:
1. Create a set called unique_candies to keep track of the unique candy types.
2. Iterate over each candy in the array candyType.
3. Add each candy to the set unique_candies.
4. After iterating over all candies, check the length of the set unique_candies and store it in a variable num_unique.
5. Calculate max_types as the minimum of num_unique and n/2, where n is the length of the input array candyType (since Alice can only eat n/2 candies).
6. Return the value of max_types.

**Time Complexity**: O(n) 

**Space Complexity**: O(1)

```python
def max_candies(candyType):
    unique_candies = set()
    for candy in candyType:
        unique_candies.add(candy)
    num_unique = len(unique_candies)
    max_types = min(num_unique, len(candyType) // 2)
    return max_types
candyType = [1, 1, 2, 2, 3, 3]
result = max_candies(candyType)
print(result)
```

**Question 3**
We define a harmonious array as an array where the difference between its maximum value
and its minimum value is exactly 1.

Given an integer array nums, return the length of its longest harmonious subsequence
among all its possible subsequences.

A subsequence of an array is a sequence that can be derived from the array by deleting some or no elements without changing the order of the remaining elements.

**Example 1:**
Input: nums = [1,3,2,2,5,2,3,7]
Output: 5

**Explanation:** The longest harmonious subsequence is [3,2,2,2,3].


`Approach`:
1. Initialize a dictionary, counter, to keep track of the frequency of each number in the input array nums.
2. Iterate through each number in nums and update the frequencies in the counter dictionary.
3. Initialize a variable max_length to store the maximum length of the harmonious subsequence found so far.
4. Iterate through each number in nums again, and for each number num, check if num + 1 exists in the counter dictionary.
- If it exists, calculate the length of the harmonious subsequence starting with num by adding the frequencies of num and num + 1 from the counter dictionary.
- Update max_length if the length of the current harmonious subsequence is greater than max_length.
5. Return max_length as the result.


**Time Complexity** : `O(n)`

**Space Complexity** : `O(n)`

```python
def findLHS(nums):
    counter = {}
    max_length = 0

    for num in nums:
        counter[num] = counter.get(num, 0) + 1

    for num in nums:
        if num + 1 in counter:
            length = counter[num] + counter[num + 1]
            max_length = max(max_length, length)

    return max_length
nums = [1, 3, 2, 2, 5, 2, 3, 7]
result = findLHS(nums)
print(result)
```

**Question 4**
You have a long flowerbed in which some of the plots are planted, and some are not.
However, flowers cannot be planted in adjacent plots.
Given an integer array flowerbed containing 0's and 1's, where 0 means empty and 1 means not empty, and an integer n, return true if n new flowers can be planted in the flowerbed without violating the no-adjacent-flowers rule and false otherwise.

**Example 1:**
Input: flowerbed = [1,0,0,0,1], n = 1

Output: true

`Approach`:
1. Initialize a variable count to keep track of the number of flowers planted.
2. Iterate through each plot in the flowerbed:
- Check if the current plot is empty (0) and its adjacent plots (if any) are also empty.
- If the conditions are satisfied, plant a new flower in the current plot by updating its value to 1 and increment the count variable by 1.
3. Check if the number of planted flowers (count) is equal to or greater than the required number of new flowers (n).
- If true, return True since all the flowers can be planted.
- If false, return False since it is not possible to plant all the flowers.

**Time Complexity** : `O(n)`

**Space Complexity** : `O(1)`

```python
def canPlaceFlowers(flowerbed, n):
    length = len(flowerbed)
    count = 0
    i = 0

    while i < length:
        if (
            flowerbed[i] == 0 and
            (i == 0 or flowerbed[i-1] == 0) and
            (i == length - 1 or flowerbed[i+1] == 0)
        ):
            flowerbed[i] = 1
            count += 1
            if count >= n:
                return True
        i += 1

    return False
flowerbed = [1, 0, 0, 0, 1]
n = 1
result = canPlaceFlowers(flowerbed, n)
print(result)
```

**Question 5**
Given an integer array nums, find three numbers whose product is maximum and return the maximum product.

**Example 1:**
Input: nums = [1,2,3]

Output: 6

`Approach`:
1. Sort the given array `nums `in ascending order.
2. Calculate two possible products:
- The product of the three largest elements in `nums`, which would be `nums[-1] * nums[-2] * nums[-3]`.
- The product of the two smallest elements (if they are negative) and the largest element, which would be `nums[0] * nums[1] * nums[-1]`.
3. Return the maximum of the two products calculated in step 2.

**Time Complexity** : `O(n log n)`

**Space Complexity** : `O(1)`

```python
def maximumProduct(nums):
    nums.sort()
    return max(nums[-1] * nums[-2] * nums[-3], nums[0] * nums[1] * nums[-1])
nums = [1, 2, 3]
result = maximumProduct(nums)
print(result)
```

**Question 6**
Given an array of integers nums which is sorted in ascending order, and an integer target,
write a function to search target in nums. If target exists, then return its index. Otherwise,
return -1.

You must write an algorithm with O(log n) runtime complexity.

**Input**: nums = [-1,0,3,5,9,12], target = 9
**Output**: 4

**Explanation**: 9 exists in nums and its index is 4

`Approach`:
1. Initialize two pointers, `left` and `right`, to the start and end of the array respectively.
2. Repeat the following steps until `left `is less than or equal to `right`:
3. Calculate the middle index as `mid = left + (right - left) // 2`.
- If the middle element `nums[mid]` is equal to the target, return mid as the index of the target.
- If the middle element is greater than the target, update `right` to `mid - 1` to search in the left half of the array.
- If the middle element is less than the target, update `left `to `mid + 1` to search in the right half of the array.
4. If the target is not found in the array, return -1 to indicate that it does not exist.

**Time Complexity** : `O(log n)`

**Space Complexity** : `O(1)`

```python
def search(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1
nums = [-1, 0, 3, 5, 9, 12]
target = 9
result = search(nums, target)
print(result)
```

**Question 7**
An array is monotonic if it is either monotone increasing or monotone decreasing.

An array nums is monotone increasing if for all i <= j, nums[i] <= nums[j]. An array nums is
monotone decreasing if for all i <= j, nums[i] >= nums[j].

Given an integer array nums, return true if the given array is monotonic, or false otherwise.

**Example 1:**

**Input**: nums = [1,2,2,3]

**Output**: true

`Approach`:
1. Initialize two flags, `isIncreasing` and `isDecreasing`, to track if the array is monotonic increasing or decreasing, respectively.
2. Iterate through the array nums from index 1 to the end:
- Check if the current element `nums[i]` is greater than the previous element `nums[i-1]`. If it is, set `isIncreasing` to True.
- Check if the current element `nums[i]` is less than the previous element `nums[i-1]`. If it is, set `isDecreasing` to True.
- If both flags are already True, we can stop the iteration early and return False, as the array cannot be both increasing and decreasing at the same time.
3. Finally, return True if either isIncreasing or isDecreasing is True. Otherwise, return False.

**Time Complexity** : `O(n)`

**Space Complexity** : `O(1)`

```python
def isMonotonic(nums):
    isIncreasing = False
    isDecreasing = False

    for i in range(1, len(nums)):
        if nums[i] > nums[i-1]:
            isIncreasing = True
        elif nums[i] < nums[i-1]:
            isDecreasing = True
        if isIncreasing and isDecreasing:
            return False

    return True
nums = [1, 2, 2, 3]
result = isMonotonic(nums)
print(result)
```

**Question 8**
You are given an integer array nums and an integer k.

In one operation, you can choose any index i where 0 <= i < nums.length and change nums[i] to nums[i] + x where x is an integer from the range [-k, k]. You can apply this operation at most once for each index i.

The score of nums is the difference between the maximum and minimum elements in nums.

Return the minimum score of nums after applying the mentioned operation at most once for each index in it.

**Example 1**:
**Input**: nums = [1], k = 0
**Output**: 0

**Explanation**: The score is max(nums) - min(nums) = 1 - 1 = 0.

`Approach`:
1. Find the maximum and minimum elements in the array `nums`.
2. If the difference between the maximum and minimum elements is less than or equal to 2*k, return 0. This is because we can perform the operation such that the maximum and minimum elements become equal, resulting in a score of 0.
3. Otherwise, calculate the potential minimum and maximum values that can be achieved after the operation. The potential minimum would be the current minimum minus k, and the potential maximum would be the current maximum plus k.
4. Return the difference between the potential maximum and potential minimum, which represents the minimum score achievable after the operation

**Time Complexity** : `O(n)`

**Space Complexity** : `O(1)`

```python
def minimumScore(nums, k):
    minimum = min(nums)
    maximum = max(nums)

    if maximum - minimum <= 2 * k:
        return 0

    potential_min = minimum + k
    potential_max = maximum - k

    return potential_max - potential_min
nums = [1]
k = 0
result = minimumScore(nums, k)
print(result)
```
